# Plot the data

1. Using inline format strings (see Notes in https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

```python
plt.plot(ages, all_dev, 'k--.', label='All Devs')
plt.plot(ages, py_dev, 'b-o', label='Python Devs')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()

plt.show()
```

2. Using atributes `color`, `linestyle` and `marker`

```python
#plt.xkcd()

plt.plot(ages, all_dev, color='k', linestyle='--', marker='.', label='All Devs')
plt.plot(ages, py_dev, color='b', marker='o', label='Python Devs')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()

plt.show()
```

3. Just plain

```python
#plt.xkcd()

plt.plot(ages, all_dev, color='k', linestyle='--', label='All Devs')
plt.plot(ages, py_dev, color='b', label='Python Devs')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()

plt.show()
```

4. Using hex values for colors

```python
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_dev, color='#5a7d9a', label='Python')
plt.plot(ages, js_dev, color='#adad3b', label='Javascript')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()

plt.grid(True)
plt.show()
```

5. Widening the lines

```python
plt.plot(ages, py_dev, color='#5a7d9a', linewidth=3, label='Python')
plt.plot(ages, js_dev, color='#adad3b', linewidth=3, label='Javascript')
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Devs')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()

plt.grid(True)
plt.show()
```

6. adjust layout

```python
plt.plot(ages, py_dev, color='#5a7d9a', linewidth=3, label='Python')
plt.plot(ages, js_dev, color='#adad3b', linewidth=3, label='Javascript')
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Devs')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline

import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')
```

Example data

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x, y)

plt.tight_layout()
plt.show()
```

Changing the markers size to 100, the color to green and markers itself to show an 'X'

```python
plt.scatter(x, y, s=100, c='green', marker='X')

plt.tight_layout()
plt.show()
```

Changing the color and the line width of the edge of the marker. The `alpha` parameter gives a transparency to the markers

```python
plt.scatter(x, y, s=100, c='green', edgecolor='black', linewidth=1, alpha=0.5)

plt.tight_layout()
plt.show()
```

Giving multiple color to each marker.

Ilustrate, for example, the degree of satisfaction of each client to some products

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.5)

plt.tight_layout()
plt.show()
```

Pasing a color map (instead of all grey)

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.5)

plt.tight_layout()
plt.show()
```

But what does the color represent? Are the light colors an higher or a lower value? And what valoe?

We need to give a color bar legend...

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
```

Changing the sizes to each marker.

Ilustrate, for example, the degree of satisfaction of each client to some products

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
```

Real world data, from CSV file

```python
data = pd.read_csv('2019-05-31-data.csv')

data
```

Split the data in lists

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

Using a log-scale for our axes, due to the outlier

```python
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

Using the `ratio` column as colors

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

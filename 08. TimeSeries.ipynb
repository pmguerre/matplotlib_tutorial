```python
%matplotlib inline

from datetime import datetime, timedelta

import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')
```

Example data

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30),
]

y = [0, 1, 3, 4, 6, 5, 7]
```

Plot example data

```python
plt.plot_date(dates, y)

plt.tight_layout()
plt.show()
```

Ploting a line, instead of just markers (if you wanted to remove the markers altogether, just pass `marker=None`

```python
plt.plot_date(dates, y, linestyle='solid')

plt.tight_layout()
plt.show()
```

Running an autoformat on the `figure` of the graph

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```

Changing the format of the date

```python
plt.plot_date(dates, y, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

Now using some real data, from a CSV file

```python
data = pd.read_csv('data08.csv')

data
```

```python
price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```

The "date" is not a real date, but a string.

To fix this we need to use a Pandas method to convert the strings to real datetime format.

As we are at it, we also sorted the values according to the date (now real date)

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
```

Now we can plot the date **and** format the axis

```python
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```

# Plot Fills

```python
from matplotlib import pyplot as plt
%matplotlib inline

import pandas as pd
```

Read the data from the CSV file

```python
data = pd.read_csv('data05.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

Plot the data

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

Using fillbetween to fill the spaces between the data

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.fill_between(ages, py_salaries)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()

plt.tight_layout()

plt.show()
```

Giving some transparency to the fill (`alpha=0.25`)

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()

plt.tight_layout()

plt.show()
```

Fill between the overall median salary and the Python salary

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, y2=overall_median, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()

plt.tight_layout()

plt.show()
```

Conditional filling

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, y2=overall_median, where=(py_salaries > overall_median),
                 interpolate=True, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, y2=overall_median,
                 where=(py_salaries > overall_median), interpolate=True,
                 alpha=0.25)

plt.fill_between(ages, py_salaries, y2=overall_median,
                 where=(py_salaries <= overall_median), interpolate=True,
                 alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()

plt.tight_layout()

plt.show()
```

with custom colors

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, y2=overall_median,
                 where=(py_salaries > overall_median), interpolate=True,
                 alpha=0.25)

plt.fill_between(ages, py_salaries, y2=overall_median,
                 where=(py_salaries <= overall_median), interpolate=True,
                 color='red', alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()

plt.tight_layout()

plt.show()
```

Fill in between two plots

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, y2=dev_salaries,
                 where=(py_salaries > dev_salaries), interpolate=True,
                 alpha=0.25, label='Above average')

plt.fill_between(ages, py_salaries, y2=dev_salaries,
                 where=(py_salaries <= dev_salaries), interpolate=True,
                 color='red', alpha=0.25, label='Bellow average')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()

plt.tight_layout()

plt.show()
```

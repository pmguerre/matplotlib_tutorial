# Pie Charts

First (as always) import the module

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

## A simple example

```python
slices = [60, 40]

plt.pie(slices)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

Adding the labels...

```python
slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

```python
The values don't need to add up to 100
```

```python
slices = [240, 160]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

Changing the color of the wedge between each slice

```python
slices = [120, 80]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

Changing the colors

```python
slices = [120, 80, 30, 20]
labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

## Now with some data

Language Popularity

```python
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097,
          23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell',
          'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

```python
plt.style.use("fivethirtyeight")

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

This many items are very hard to read. Pie Charts should be used for about 5 items...

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.style.use("fivethirtyeight")

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Emphatize a item

```python
explode = [0, 0, 0, 0.1, 0]    # Explode the 4th element 10% of the radious

plt.style.use("fivethirtyeight")

plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
explode = [0, 0, 0, 0.5, 0]    # Explode the 4th element 50% of the radious

plt.style.use("fivethirtyeight")

plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

Some more options

```python
explode = [0, 0, 0, 0.1, 0]    # Explode the 4th element 50% of the radious

plt.style.use("fivethirtyeight")

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90,
        autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
